```python
import html2text
from deep_translator import (GoogleTranslator,MicrosoftTranslator,MyMemoryTranslator)
import pandas as pd
from transformers import AutoTokenizer, AutoModelWithLMHead, pipeline
import requests
from bs4 import BeautifulSoup
import bleach

def get_text(url):

    rs = requests.get(url)
    url_text = rs.text

    title_name = url_text.split('<TITLE>')[1].split('</TITLE>')[0]
    text = url_text.split('"articleBody":')[1:][0][1:].split('"')[0]

    h = html2text.HTML2Text()
    h.ignore_links = True
    text1 = h.handle(text)

    text = bleach.clean(text1, tags=[], strip=True)

    return [title_name,text]
```
